```python
import pandas as pd

file_path = "/content/TheHackerNews_Dataset.xlsx"  # Update path if needed
df = pd.read_excel(file_path)

# Check dataset
df.head()
```

```python
!pip uninstall -y nltk
!pip install --upgrade nltk
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
import re



# Function to clean text
def clean_text(text):
    text = re.sub(r'[^a-zA-Z0-9]', ' ', str(text))  # Remove special characters
    text = text.lower()  # Lowercase
    words = text.split()
    words = [word for word in words if word not in stop_words]  # Remove stopwords
    return " ".join(words)

# Apply to dataset
df['clean_title'] = df['Title'].astype(str).apply(clean_text)
df[['Title', 'clean_title']].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X_bow = vectorizer.fit_transform(df['clean_title'])

print("Shape of BoW matrix:", X_bow.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(df['clean_title'])

print("Shape of TF-IDF matrix:", X_tfidf.shape)
```

```python
ngram_vectorizer = CountVectorizer(ngram_range=(1, 2))  # Unigram + Bigram
X_ngram = ngram_vectorizer.fit_transform(df['clean_title'])

print("Shape of N-gram matrix:", X_ngram.shape)
```
